```
┌─────────────────────────────────────────────┐
│               CLIENT LAYER                  │

│  │  Expo App    │    │  Vue 2 + Vuetify  │  │

│  └──────┬───────┘    └────────┬──────────┘  │

          ▼                   ▼
┌─────────────────────────────────────────────┐
│               API LAYER                     │
│         Node.js + Express.js                │
│   JWT Auth · Route Controllers · Services   │
└──────────────────┬──────────────────────────┘
                   │

┌─────────────────────────────────────────────┐
│              DATA LAYER                     │
```

### Key Design Decisions

| Decision | Rationale |
|---|---|
| Firestore over MySQL | Native Firebase ecosystem; seamless deployment to Cloud Functions & Hosting; auto-scaling |
| Expo over bare React Native | Faster MVP iteration; OTA updates |
| JWT with rotation | Stateless auth; tokens refreshed on expiry |
| Vue 2 + Vuetify for admin | Mature component library; quick UI scaffolding |
| Monorepo | Shared types/utilities; single clone to get full stack running |

## 8. Firestore Data Model

Firestore is a **NoSQL document database**. Data is organized into **collections** containing **documents**.  
There are no tables, no joins, and no fixed schema — each document can have different fields.

### Collections Overview

```
Firestore (manabisai)
├── categories/              # Food category documents
│   └── {categoryId}
│       └── { name: "Nasi Katok" }
│
├── users/                   # User account documents
│   └── {userId}
│       └── {
│             name, email, password (bcrypt),
│             role: "user" | "admin", created_at
│           }
│
├── spots/                   # Food location documents
│   └── {spotId}
│       └── {
│             name, description, latitude, longitude,
│             address, category_id (ref), category (name),
│             is_halal, created_at
│           }
│
├── reviews/                 # Review documents
│   └── {reviewId}
│       └── {
│             user_id (ref), spot_id (ref),
│             rating: 1-5, comment, created_at
│           }
│
└── refresh_tokens/          # JWT refresh token documents
    └── {tokenId}
        └── { user_id (ref), token, expires_at }
```

### Document Schemas

#### `users` collection
| Field | Type | Notes |
|---|---|---|
| `name` | string | Required |
| `email` | string | Required, unique (enforced in app logic) |
| `password` | string | bcrypt hash (salt rounds: 10) |
| `role` | string | `"user"` or `"admin"` (default: `"user"`) |
| `created_at` | string | ISO 8601 timestamp |

#### `spots` collection
| Field | Type | Notes |
|---|---|---|
| `name` | string | Required |
| `description` | string | Optional |
| `latitude` | number | Decimal degrees |
| `longitude` | number | Decimal degrees |
| `address` | string | Optional |
| `category_id` | string | Firestore doc ID from `categories` |
| `category` | string | Denormalized category name for read efficiency |
| `is_halal` | boolean | Default: `true` |
| `created_at` | string | ISO 8601 timestamp |

#### `categories` collection
| Field | Type | Notes |
|---|---|---|
| `name` | string | Required, unique |

#### `reviews` collection
| Field | Type | Notes |
|---|---|---|
| `user_id` | string | Firestore doc ID from `users` |
| `spot_id` | string | Firestore doc ID from `spots` |
| `rating` | number | 1–5 |
| `comment` | string | Optional |
| `created_at` | string | ISO 8601 timestamp |

#### `refresh_tokens` collection
| Field | Type | Notes |
|---|---|---|
| `user_id` | string | Firestore doc ID from `users` |
| `token` | string | JWT refresh token string |
| `expires_at` | string | ISO 8601 expiry time |

### Denormalization Strategy

Unlike the previous MySQL 3NF schema, Firestore benefits from **controlled denormalization**:  
- The `spots` collection stores the `category` name alongside `category_id` to avoid a second read when displaying spots.  
- Average ratings are computed on-read by querying the `reviews` collection (no counter document — acceptable for MVP scale).  

### Geo Queries

Firestore does **not** natively support geo-distance queries. The current implementation:  
1. Fetches all spots from Firestore  
2. Filters by Haversine distance in JavaScript  

This works fine for small datasets (<1000 spots). For production scale, consider:  
- **Geohash indexing** (e.g. `geofirestore` library)  
- **Algolia Places** or **Google Places API** for geo search

## 9. Authentication Flow (JWT)

```
Client                        API Server                  Firestore
  │                               │                          │
  │── POST /auth/login ──────────▶│                          │
  │   { email, password }         │── Query users by email ─▶│
  │                               │◀─ user doc ──────────────│
  │                               │  bcrypt.compare()
  │                               │  sign accessToken (15m)
  │                               │  sign refreshToken (7d)
  │                               │── ADD refresh_tokens ────▶│
  │◀─ { accessToken, refresh } ───│                          │
  │                               │                          │
  │── GET /spots (Bearer token) ─▶│                          │
  │                               │  verify accessToken
  │◀─ spots data ─────────────────│                          │
  │                               │                          │
  │── POST /auth/refresh ────────▶│                          │
  │   { refreshToken }            │── Query refresh_tokens ─▶│
  │                               │  verify + issue new pair │
  │◀─ { newAccessToken } ─────────│                          │
```

### Rules
- Access tokens expire in **15 minutes**
- Refresh tokens expire in **7 days** and are stored in the `refresh_tokens` collection
- On logout, the refresh token document is **deleted** from Firestore
- Passwords are hashed with **bcrypt** (salt rounds: 10)

## 10. API Reference

Base URL: `http://localhost:3000/api`  
All protected routes require: `Authorization: Bearer <accessToken>`

### Auth
| Method | Endpoint | Auth | Description |
|---|---|---|---|
| POST | `/auth/register` | ❌ | Register new user |
| POST | `/auth/login` | ❌ | Login, returns token pair |
| POST | `/auth/refresh` | ❌ | Refresh access token |
| POST | `/auth/logout` | ✅ | Invalidate refresh token |

### Spots
| Method | Endpoint | Auth | Description |
|---|---|---|---|
| GET | `/spots` | ❌ | List all spots |
| GET | `/spots/:id` | ❌ | Get single spot |
| GET | `/spots/nearby` | ❌ | Get spots by lat/lng radius |
| POST | `/spots` | ✅ Admin | Create new spot |
| PUT | `/spots/:id` | ✅ Admin | Update spot |
| DELETE | `/spots/:id` | ✅ Admin | Delete spot |

### Reviews
| Method | Endpoint | Auth | Description |
|---|---|---|---|
| GET | `/spots/:id/reviews` | ❌ | Get reviews for a spot |
| POST | `/spots/:id/reviews` | ✅ | Submit a review |
| DELETE | `/reviews/:id` | ✅ | Delete own review |

### Categories
| Method | Endpoint | Auth | Description |
|---|---|---|---|
| GET | `/categories` | ❌ | List all categories |

## 11. Deployment (Firebase)

The app is now fully aligned with the Firebase ecosystem.

| Service | Current | Planned |
|---|---|---|
| API | localhost:3000 | Firebase Cloud Functions (or Cloud Run) |
| Admin Panel | localhost:8080 | Firebase Hosting |
| Database | Firestore (Firebase) | Same — already configured |
| Auth tokens | Firestore collection | Firebase Auth (optional future upgrade) |

### Deployment Steps

1. **Install Firebase CLI**: `npm install -g firebase-tools`
2. **Login**: `firebase login`
3. **Initialize** (if not done): `firebase init`  
   - Select **Hosting** and **Functions**
4. **Deploy admin panel**: build with `npm run build` in `admin/`, point hosting to `admin/dist/`
5. **Deploy API**: wrap Express app in Cloud Functions (`functions.https.onRequest(app)`)
6. **Set env vars**: `firebase functions:config:set jwt.secret="..."` etc.
7. **Deploy**: `firebase deploy`

### Firestore Security Rules

Before going public, add Firestore security rules to protect collections:
```
rules_version = '2';
service cloud.firestore {
  match /databases/{database}/documents {
    // Public read for spots and categories
    match /spots/{spotId} { allow read: if true; }
    match /categories/{categoryId} { allow read: if true; }

    // Reviews: anyone can read, authenticated users can write
    match /reviews/{reviewId} {
      allow read: if true;
      allow create: if request.auth != null;
      allow delete: if request.auth != null
                    && resource.data.user_id == request.auth.uid;
    }

    // Users: no direct client access (managed via API)
    match /users/{userId} { allow read, write: if false; }

    // Refresh tokens: no direct client access
    match /refresh_tokens/{tokenId} { allow read, write: if false; }
  }
}
```
